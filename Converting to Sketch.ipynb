# Importing Libraries

```python
import cv2
import matplotlib.pyplot as plt
plt.style.use("seaborn")
%matplotlib inline
```

# Original Image

```python
image=cv2.imread(r"D:\Users\user\puppy.jpg")
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,8))
plt.imshow(image)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

# Gray-Scale Image

```python
gray_img=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,8))
plt.imshow(gray_img)
plt.axis("off")
plt.title("Gray-scale Image")
plt.show()
```

# Inverting Image

```python
invt_img=cv2.bitwise_not(gray_img)
plt.figure(figsize=(8,8))
plt.imshow(invt_img,cmap="gray")
plt.axis("off")
plt.title("Inverted Gray-scale Image")
plt.show()
```

# Blur Image

```python
img_smoothing = cv2.GaussianBlur(invt_img, (21, 21),sigmaX=0, sigmaY=0)
plt.figure(figsize=(8,8))
plt.imshow(img_smoothing,cmap="gray")
plt.axis("off")
plt.title("Smoothen Image")
plt.show()
```

# Final sketch image

```python
final = cv2.divide(gray_img, 255 - img_smoothing, scale=255)
plt.figure(figsize=(8,8))
plt.imshow(final,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```
